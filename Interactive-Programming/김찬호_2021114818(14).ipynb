#### 판다스를 이용한 평균, 최소값, 최대값, 총합계 구하기
  * 아래 주어진 df를 이용하여 평균, 최소값, 최대값, 총합계 구하기
  * 계산한 결과를 엑셀 파일로 저장(to_excel)하고 확인(read_excel)하기
----

```python
import pandas as pd
import random

random.seed(42)

name_00 = pd.Series(["강하나", "김대기", "박식해", "정타로", "하나더",  
                     "항공호", "이기대", "소박해", "대단해", "성공해"], name="이름")
math_01 = pd.Series([x for x in random.sample(range(60, 100), 10)], name="수치해석")
math_02 = pd.Series([x for x in random.sample(range(60, 100), 10)], name="해석학")
stat_03 = pd.Series([x for x in random.sample(range(60, 100), 10)], name="확률론")
matx_04 = pd.Series([x for x in random.sample(range(60, 100), 10)], name="선형대수")
lang_05 = pd.Series([x for x in random.sample(range(60, 100), 10)], name="파이썬")

df = pd.concat([name_00, math_01, math_02, stat_03, matx_04, lang_05], axis=1)

print(type(df))
print(df)

df_copy = df.copy()

df_copy["평균값"] = df_copy.iloc[:, [1,2,3,4,5]].mean(axis=1)
df_copy["최소값"] = df_copy.iloc[:, [1,2,3,4,5]].min(axis=1)
df_copy["최대값"] = df_copy.iloc[:, [1,2,3,4,5]].max(axis=1)
df_copy["총합계"] = df_copy.iloc[:, [1,2,3,4,5]].sum(axis=1)

df_copy.to_excel("이름_학번.xlsx", index=False, encoding="utf-8")
dfx = pd.read_excel("이름_학번.xlsx")
dfx
```
