# 13주차 과제 2차방정식 그래프 그리기

## $y = x^2$ where $x = (-5, +5)$
## $y = -x^2$ where $x = (-5, +5)$
## $y = x \cdot 5$ where $x = (-5, +5)$
## $y = -x \cdot 5$ where $ x = (-5, +5)$
----

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

fig = plt.figure(figsize=(12, 6))            

x = np.linspace(-5, 5, 100) 

plt.title("4 diffrent graphs")

plt.xlim(-10, 10)
plt.ylim(-25, 25)

plt.xlabel(" $x$ axis")
plt.ylabel(" $y$ axis")

###############################################
# 하기 4개의 plot 함수를 이용하여 과제를 완성하기
plt.plot(x, x**2)
plt.plot(x, -x**2)
plt.plot(x, 5*x)
plt.plot(x, -5*x)
###############################################

plt.grid()      
plt.show()     
```
